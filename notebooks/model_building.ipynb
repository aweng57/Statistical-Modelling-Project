```python
import pandas as pd 
import numpy as np
import json
```

Provide model output and an interpretation of the results. 

```python
with open('df_final', 'r') as e: 
    data = json.load(e)
df_final = pd.DataFrame(data)
df_final
```

```python
df_final.describe()
```

```python
import statsmodels.api as sm
X = df_final['num_restaurant']
y = df_final['bikes']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

Initally I would assume that the number of restaurant would be a strong predicator of the number of bikes rented. However, statistically I was proven wrong. First, the R-square is 0.002 which means there is only a small portion of variability in "bikes rented" being variable beign explained by "number of restaurant" variable. This is also supported by the adjust R-squared value which is also very close to zero. Second, the P-value is 0.294 which is larger than the regular value of 0.05. this suggests that the linear relationship between "bikes rented" and "number of restaurant" is not significant. 

```python
X = df_final['num_entertainment']
y = df_final['bikes']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

Out of all the three independent variables, the number of entertainment suggests reelatively stronger relationship with the number of bike rentals. Although the R-squared value is only 0.033, which means only 3.3% of the variation in the number of bikes rented can be explained by the number of entertaiment, but the result has a P-value of 2.00e -05. This indicates that there is a significant relationship between the number of entertainment and the number of bike rentals.  

```python
X = df_final['num_playground']
y = df_final['bikes']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

The reason I chose playground as one of my independent variable is because I would assume that people usually take light transportation to nearby playgrounds. However, the result produced suggests otherwise. For example, the r-squared and adjuted r-squared values are extremely low, which means the nubmer of playgrounds is a weak predictor of the number of bike rentals. Furthermore, P-value is around 0.098 which also suggests that it is not statistically significant since it is greater than the 0.05 alpha value. In conlcusion, the model does not provide a good fit to the data.
