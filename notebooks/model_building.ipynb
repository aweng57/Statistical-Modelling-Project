```python
import pandas as pd 
import numpy as np
import json
```

Provide model output and an interpretation of the results. 

```python
with open('df_final', 'r') as e: 
    data = json.load(e)
df_final = pd.DataFrame(data)
df_final
```

```python
df_final.describe()
```

```python
import statsmodels.api as sm
X = df_final['num_restaurant']
y = df_final['bikes']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python
X = df_final['rating_x']
y = df_final['bikes']
X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```
